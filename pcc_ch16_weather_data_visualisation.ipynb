```python
"""Plotting data in a temperature chart"""

import csv # import csv from python standard library
import matplotlib.pyplot as plt # import pyplot for generating charts

filename = 'C:/Users/austi/Documents/pythonWork/jupyter notebook files/' \
           'weather data visualisation/data/sitka_weather_07-2018_simple.csv' # pathway of data file

with open(filename) as f: # open file as f
    reader = csv.reader(f) # create reader object with csv reader
    header_row = next(reader) # get first line in the file which contains the headers

    """get the high temperatures from this file"""
    highs = [] # create empty list to store high temps
    for row in reader: # for each row of data in reader file
        high = int(row[5]) # take data from tmax column and convert to integer
        highs.append(high) # add that data to the highs list
    
    """plot the high temperatures"""
    plt.style.use('dark_background') # set graph style
    fig, ax = plt.subplots(figsize=(20, 10)) # create figure and axis with pyplot, set chart window size
    ax.plot(highs, c='red') # create plot with highs list and set the color of the plot line
    
    """format plot"""
    ax.set_title("Daily high temperatures, July 2018", fontsize=24) # set chart title
    ax.set_xlabel('Day of the month', fontsize=16) # set label of x axis
    ax.set_ylabel('Temperature (F), fontsize=16') # set label of y axis
    ax.tick_params(axis= 'both', which='major', labelsize=16) # set parameters of tick marks
    
    plt.show() # generate the chart window
```

```python
"""Adding dates to x axis"""

import csv # import csv from python standard library
from datetime import datetime # import datetime module
import matplotlib.pyplot as plt # import pyplot for generating charts

filename = 'C:/Users/austi/Documents/pythonWork/jupyter notebook files/' \
           'weather data visualisation/data/sitka_weather_07-2018_simple.csv' # pathway of data file

with open(filename) as f: # open file as f
    reader = csv.reader(f) # create reader object with csv reader
    header_row = next(reader) # get first line in the file which contains the headers

    """get the high temperatures from this file"""
    dates, highs = [], [] # create empty lists to store dates and high temps
    for row in reader: # for each row of data in reader file
        current_date = datetime.strptime(row[2], '%Y-%m-%d') # format date data from column 2 in row
                                                             # and store it in a variable
        high = int(row[5]) # take data from tmax column and convert to integer
        dates.append(current_date) # add current_date to list of dates
        highs.append(high) # add that data to the highs list
    
    """plot the high temperatures"""
    plt.style.use('dark_background') # set graph style
    fig, ax = plt.subplots(figsize=(20, 10)) # create figure and axis with pyplot, set chart window size
    ax.plot(dates, highs, c='red') # create plot with dates and highs list
                                   # and set the color of the plot line
    
    """format plot"""
    ax.set_title("Daily high temperatures, July 2018", fontsize=24) # set chart title
    ax.set_xlabel('Day of the month', fontsize=16) # set label of x axis
    fig.autofmt_xdate() # auto format the dates so they dont overlap on the chart
    ax.set_ylabel('Temperature (F), fontsize=16') # set label of y axis
    ax.tick_params(axis= 'both', which='major', labelsize=16) # set parameters of tick marks
    
    plt.show() # generate the chart window
```

```python
"""plotting rainfall data for alaska chart"""

import csv # import csv from python standard library
from datetime import datetime # import datetime module
import matplotlib.pyplot as plt # import pyplot for generating charts

filename = 'C:/Users/austi/Documents/pythonWork/jupyter notebook files/' \
           'weather data visualisation/data/sitka_weather_2018_simple.csv' # pathway of data file

with open(filename) as f: # open file as f
    reader = csv.reader(f) # create reader object with csv reader
    header_row = next(reader) # get first line in the file which contains the headers

    """get the dates, high and low temperatures, and rainfall from this file"""
    dates, highs, lows, rainfall = [], [], [], [] # create empty lists to store dates, high temps,
                                              # low temps, and rainfall
    for row in reader: # for each row of data in reader file
        current_date = datetime.strptime(row[2], '%Y-%m-%d') # format date data from column 2 in row
                                                             # and store it in a variable
        high = int(row[5]) # take data from tmax column and convert to integer
        low = int(row[6]) # take data from tmin column and convert to integer
        rain = float(row[3]) * 10 # take data from prcp column and convert to integer
        dates.append(current_date) # add current_date to list of dates
        highs.append(high) # add that data to the highs list
        lows.append(low) # add to list of lows
        rainfall.append(rain) # add to list of rainfall
    
    """plot the high and low temperatures"""
    plt.style.use('dark_background') # set graph style
    fig, ax = plt.subplots(figsize=(20, 10)) # create figure and axis with pyplot, set chart window size
    ax.plot(dates, highs, c='red', alpha=0.5) # create plot with dates and highs list,
                                              # set the color of the plot line, and opacity
    ax.plot(dates, lows, c='cyan', alpha=0.5) # create plot of dates, lows list,
                                              # set color to cyan and oppacity
    ax.plot(dates, rainfall, c='green') # create plot line for rainfall
    ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1) # fill in between the plotted lines
    
    """format plot"""
    ax.set_title("Daily high and low temperatures - 2018", fontsize=24) # set chart title
    ax.set_xlabel('Day of the month', fontsize=16) # set label of x axis
    fig.autofmt_xdate() # auto format the dates so they dont overlap on the chart
    ax.set_ylabel('Temperature (F)\nRainfall(inches*10)', fontsize=16) # set label of y axis
    ax.tick_params(axis= 'both', which='major', labelsize=16) # set parameters of tick marks
    
    plt.show() # generate the chart window
```

```python
"""combining both temp graphs together"""

import csv # import csv from python standard library
from datetime import datetime # import datetime module
import matplotlib.pyplot as plt # import pyplot for generating charts

filename = 'C:/Users/austi/Documents/pythonWork/jupyter notebook files/' \
           'weather data visualisation/data/sitka_weather_2018_simple.csv' # pathway of data file

with open(filename) as f: # open file as f
    reader = csv.reader(f) # create reader object with csv reader
    header_row = next(reader) # get first line in the file which contains the headers

    """sitka alaska data"""
    dates, highs, lows = [], [], [] # create empty lists to store dates, high temps, and low temps
    for row in reader: # for each row of data in reader file
        current_date = datetime.strptime(row[2], '%Y-%m-%d') # format date data from column 2 in row
                                                             # and store it in a variable
        high = int(row[5]) # take data from tmax column and convert to integer
        low = int(row[6]) # take data from tmin column and convert to integer
        dates.append(current_date) # add current_date to list of dates
        highs.append(high) # add that data to the highs list
        lows.append(low) # add to list of lows
    
"""death valley data"""
filename1 = 'C:/Users/austi/Documents/pythonWork/jupyter notebook files/' \
           'weather data visualisation/data/death_valley_2018_simple.csv' # pathway of data file

with open(filename1) as f1: # open file as f
    reader1 = csv.reader(f1) # create reader object with csv reader
    header_row1 = next(reader1) # get first line in the file which contains the headers

    """get the dates and high and low temperatures from this file"""
    dates1, highs1, lows1 = [], [], [] # create empty lists to store dates, high temps, and low temps
    for row1 in reader1: # for each row of data in reader file
        current_date1 = datetime.strptime(row1[2], '%Y-%m-%d') # format date data from column 2 in row
                                                             # and store it in a variable
        try: # attempt to read data
            high1 = int(row1[4]) # take data from tmax column and convert to integer
            low1 = int(row1[5]) # take data from tmin column and convert to integer
        except ValueError: # if data is missing
            print(f"Missing data for {current_date}") # print this message to terminal
        else: # continue with loop
            dates1.append(current_date1) # add current_date to list of dates
            highs1.append(high1) # add that data to the highs list
            lows1.append(low1) # add to list of lows
    
"""plot the high and low temperatures for each data set"""
plt.style.use('dark_background') # set graph style
fig, ax = plt.subplots(figsize=(20, 10)) # create figure and axis with pyplot, set chart window size

# sitka plot
ax.plot(dates, highs, c='cyan', alpha=0.5) # create plot with dates and highs list,
                                              # set the color of the plot line, and opacity
ax.plot(dates, lows, c='blue', alpha=0.5) # create plot of dates, lows list,
                                              # set color to cyan and oppacity
ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1) # fill in between the plotted lines

# death valley plot
ax.plot(dates1, highs1, c='red', alpha=0.5) # create plot with dates1 and highs1 list,
                                              # set the color of the plot line, and opacity
ax.plot(dates1, lows1, c='yellow', alpha=0.5) # create plot of dates1, lows1 list,
                                              # set color to cyan and oppacity
ax.fill_between(dates1, highs1, lows1, facecolor='orange', alpha=0.1) # fill in between the plotted lines
    
"""format plot"""
ax.set_title("Daily high and low temperatures\nDeath Valley, California(reds), Sitka, Alaska(blues) - 2018", fontsize=24) # set chart title
ax.set_xlabel('Month', fontsize=16) # set label of x axis
fig.autofmt_xdate() # auto format the dates so they dont overlap on the chart
ax.set_ylabel('Temperature (F)', fontsize=16) # set label of y axis
ax.tick_params(axis= 'both', which='major', labelsize=16) # set parameters of tick marks
    
plt.show() # generate the chart window
```
